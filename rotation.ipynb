##### rotaion using matrix

```python
import cv2
image=cv2.imread('images/lenna.png')
center= (image.shape[1]//2,image.shape[0]//2)
angle=-67
rotation_matrix=cv2.getRotationMatrix2D(center,angle,1.0)
rotation_image=cv2.warpAffine(image,rotation_matrix,(image.shape[1],image.shape[0]))
cv2.imshow('Rotated Image', rotation_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##change rotaion point
import cv2
image=cv2.imread('images/lenna.png')

angle=-67
rotation_matrix=cv2.getRotationMatrix2D((100,100),angle,1.0)
rotation_image=cv2.warpAffine(image,rotation_matrix,(image.shape[1],image.shape[0]))
cv2.imshow('Rotated Image', rotation_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```
